```python
data = pd.read_csv('/content/opsd_germany_daily.csv', index_col=0, parse_dates=True)

data['Year'] = data.index.year
data['Month'] = data.index.month
data['Day'] = data.index.day_name()
data.sample(5, random_state=0)
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(11, 4)})
data['Consumption'].plot(linewidth=0.5)
```

```python
cols_plot = ['Consumption', 'Solar', 'Wind']
axes = data[cols_plot].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11,9), subplots=True)
for ax in axes:
  ax.set_ylabel('Daily Totals (GWh)')
plt.show()
```

seasonality

```python
fig, axes = plt.subplots(3, 1, figsize=(11,10), sharex=True)
for name, ax in zip(['Consumption', 'Solar', 'Wind'], axes):
  sns.boxplot(data=data, x='Month', y=name, ax=ax)
  ax.set_ylabel('GWh')
  ax.set_title(name)
plt.show()
```

```python
data_cols = ['Consumption', 'Wind', 'Solar']
data_week = data[data_cols].resample('W').mean()
data_week.head(3)
```

```python
print(data.shape[0])
print(data_week.shape[0])
```

```python
start, end = '2017-01', '2017-12'
fig, ax = plt.subplots()
ax.plot(data.loc[start:end, 'Solar'], marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(data_week.loc[start:end, 'Solar'], marker='o', markersize=8, linestyle='-', label='Weekly')
ax.legend()
plt.show()
```

```python
data_annual = data[data_cols].resample('Y').sum(min_count=360)
data_annual['Year'] = data_annual.index.year
data_annual.set_index('Year', inplace=True)
data_annual['Wind+solar/consumption'] = data_annual['Wind+Solar'] / data_annual['Consumption']
data_annual.tail(3)
```

```python
ax = data_annual.loc[2012: , 'Wind+solar/consumption'].plot.bar()
ax.set_ylabel('Wind+Solar/Consumption')
plt.show()
```

```python
data_7d = data[data_cols].rolling(7, center=True).mean()
data_7d.head(10)
```

```python
import matplotlib.dates as mdates

data_365d = data[data_cols].rolling(window=365, center=True, min_periods=360).mean()

fig, ax = plt.subplots()
ax.plot(data['Consumption'], marker='.', markersize=2, color='0.6', linestyle='None', label='Consumption')
ax.plot(data_7d['Consumption'], linewidth=2, label='7d Rolling Mean')
ax.plot(data_365d['Consumption'], color='0.2', linewidth=3, label='365d Rolling Mean')
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
for name in ['Wind', 'Solar', 'Wind+Solar']:
  ax.plot(data_365d[name], linewidth=2, label=name)
ax.set_ylabel('Consumption (GWh)')
ax.legend()
plt.show()
```
